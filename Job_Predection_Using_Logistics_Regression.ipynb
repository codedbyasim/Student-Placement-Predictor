```python
# Import Libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Load Dataset
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# Encode target
le = LabelEncoder()
df['PlacementStatus'] = le.fit_transform(df['PlacementStatus'])
```

```python
df.head()
```

```python
# Select relevant features
features = ['Internships', 'CGPA', 'Workshops/Certifications', 'AptitudeTestScore', 'SoftSkillsRating']
X = df[features]
y = df['PlacementStatus']
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

# Take user input
internships = int(input("Enter number of internships: "))
cgpa = float(input("Enter CGPA: "))
workshops = int(input("Enter number of workshops/certifications: "))
aptitude_score = int(input("Enter aptitude test score (0–100): "))
soft_skills = float(input("Enter soft skills rating (e.g., 4.2): "))

# Create DataFrame for input
user_input = pd.DataFrame([{
    'Internships': internships,
    'CGPA': cgpa,
    'Workshops/Certifications': workshops,
    'AptitudeTestScore': aptitude_score,
    'SoftSkillsRating': soft_skills
}])

# Apply the same scaler used on training data
user_input_scaled = scaler.transform(user_input)

# Predict
prediction = model.predict(user_input_scaled)[0]
label = "Placed" if prediction == 1 else "Not Placed"

print("\nPrediction for the given student:", label)
```
